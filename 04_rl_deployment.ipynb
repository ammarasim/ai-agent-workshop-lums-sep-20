**Adaptive Q-Learning Agent**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time

class QLearningRobot:
    def __init__(self, grid_size=3):
        self.grid_size = grid_size
        self.q_table = np.zeros((grid_size, grid_size, 4))  # 4 actions: up, right, down, left

        # Initialize rewards with distance-based rewards
        self.rewards = np.full((grid_size, grid_size), -0.1)  # Cost for movement
        self.rewards[0, 0] = 0  # Start position
        self.rewards[2, 2] = 10  # Goal position (increased reward)
        self.rewards[1, 1] = -5  # Obstacle (increased penalty)

        # Add distance-based rewards to guide the robot
        goal_pos = (2, 2)
        for i in range(grid_size):
            for j in range(grid_size):
                if (i, j) != goal_pos and self.rewards[i, j] == -0.1:
                    # Manhattan distance to goal
                    distance = abs(i - goal_pos[0]) + abs(j - goal_pos[1])
                    # Closer positions get slightly better rewards
                    self.rewards[i, j] = -0.1 + (1.0 / (distance + 1)) * 0.2

        # Q-Learning parameters
        self.learning_rate = 0.1
        self.discount_factor = 0.95
        self.epsilon = 0.3  # Higher initial exploration
        self.epsilon_decay = 0.995
        self.epsilon_min = 0.05  # Higher minimum exploration

        # Actions: 0=up, 1=right, 2=down, 3=left
        self.actions = [(-1, 0), (0, 1), (1, 0), (0, -1)]
        self.action_names = ['‚Üë', '‚Üí', '‚Üì', '‚Üê']

        # Training history
        self.rewards_history = []
        self.path_history = []
        self.current_path = []

    def get_next_action(self, state):
        """Choose action using epsilon-greedy policy"""
        if np.random.random() < self.epsilon:
            return np.random.randint(4)  # Explore
        q_values = self.q_table[state[0], state[1]]
        # Add small random noise to break ties
        return np.argmax(q_values + np.random.random(4) * 1e-6)  # Exploit

    def get_next_state(self, state, action):
        """Get next state based on current state and action"""
        next_row = max(0, min(self.grid_size - 1, state[0] + self.actions[action][0]))
        next_col = max(0, min(self.grid_size - 1, state[1] + self.actions[action][1]))
        return (next_row, next_col)

    def is_valid_state(self, state):
        """Check if state is valid (not an obstacle)"""
        return self.rewards[state[0], state[1]] != -5

    def print_grid(self, current_state=None, path=None, title="Grid World"):
        """Print a text-based representation of the grid"""
        print(f"\n{title}")
        print("=" * 30)

        grid_display = [['.' for _ in range(self.grid_size)] for _ in range(self.grid_size)]

        # Mark special cells
        grid_display[0][0] = 'S'  # Start
        grid_display[2][2] = 'G'  # Goal

        # Mark obstacles
        for i in range(self.grid_size):
            for j in range(self.grid_size):
                if self.rewards[i, j] == -5:
                    grid_display[i][j] = 'X'

        # Mark path
        if path:
            for step, (r, c) in enumerate(path[:-1]):
                if grid_display[r][c] not in ['S', 'G', 'X']:
                    grid_display[r][c] = str(step + 1)

        # Mark current robot position
        if current_state:
            r, c = current_state
            if grid_display[r][c] not in ['G']:  # Don't overwrite goal
                grid_display[r][c] = 'ü§ñ'

        # Print the grid
        for row in grid_display:
            print(' | '.join(f'{cell:>2}' for cell in row))
        print()

    def visualize_q_table_text(self):
        """Print Q-table in text format"""
        print("\nQ-Table (Best Actions and Values):")
        print("=" * 40)

        for i in range(self.grid_size):
            for j in range(self.grid_size):
                if self.rewards[i, j] != -5:  # Don't show obstacles
                    best_action = np.argmax(self.q_table[i, j])
                    max_q = np.max(self.q_table[i, j])
                    print(f"({i},{j}): {self.action_names[best_action]} ({max_q:.3f})", end="  ")
                else:
                    print(f"({i},{j}): X (obstacle)", end="  ")
            print()
        print()

    def create_visual_plot(self, current_state=None, path=None, episode=0, show_plot=True):
        """Create a matplotlib visualization"""
        try:
            # Ensure episode is an integer
            if isinstance(episode, str):
                episode = len(self.rewards_history) if self.rewards_history else 0

            # Set up the plot
            fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
            fig.suptitle(f'Q-Learning Robot Navigation - Episode {episode}', fontsize=14, fontweight='bold')

            # Plot 1: Grid World
            ax1.clear()
            ax1.set_xlim(-0.5, self.grid_size - 0.5)
            ax1.set_ylim(-0.5, self.grid_size - 0.5)
            ax1.set_aspect('equal')
            ax1.set_title('Grid World with Robot', fontweight='bold')

            # Draw grid
            for i in range(self.grid_size + 1):
                ax1.axhline(i - 0.5, color='black', linewidth=1)
                ax1.axvline(i - 0.5, color='black', linewidth=1)

            # Color cells based on rewards
            for i in range(self.grid_size):
                for j in range(self.grid_size):
                    if (i, j) == (0, 0):  # Start
                        color = 'lightblue'
                        ax1.text(j, self.grid_size-1-i, 'START', ha='center', va='center', fontweight='bold', fontsize=8)
                    elif (i, j) == (2, 2):  # Goal
                        color = 'lightgreen'
                        ax1.text(j, self.grid_size-1-i, 'GOAL', ha='center', va='center', fontweight='bold', fontsize=8)
                    elif self.rewards[i, j] == -5:  # Obstacle
                        color = 'red'
                        ax1.text(j, self.grid_size-1-i, 'X', ha='center', va='center', fontweight='bold', fontsize=14, color='white')
                    else:
                        color = 'white'

                    rect = patches.Rectangle((j-0.5, self.grid_size-1-i-0.5), 1, 1,
                                           linewidth=1, edgecolor='black', facecolor=color, alpha=0.7)
                    ax1.add_patch(rect)

            # Draw path
            if path and len(path) > 1:
                for k in range(len(path) - 1):
                    start = path[k]
                    end = path[k + 1]
                    if start != end:
                        ax1.arrow(start[1], self.grid_size-1-start[0],
                                end[1] - start[1], -(end[0] - start[0]),
                                head_width=0.08, head_length=0.08, fc='blue', ec='blue', alpha=0.6)

            # Draw robot
            if current_state:
                robot = plt.Circle((current_state[1], self.grid_size-1-current_state[0]),
                                 0.2, color='orange', zorder=5)
                ax1.add_patch(robot)
                ax1.text(current_state[1], self.grid_size-1-current_state[0], 'ü§ñ',
                       ha='center', va='center', fontsize=12, zorder=6)

            ax1.set_xticks(range(self.grid_size))
            ax1.set_yticks(range(self.grid_size))
            ax1.invert_yaxis()

            # Plot 2: Q-Values
            ax2.clear()
            q_max = np.max(self.q_table, axis=2)
            im = ax2.imshow(q_max, cmap='viridis')
            ax2.set_title('Q-Values (Max per State)', fontweight='bold')

            # Add text annotations
            for i in range(self.grid_size):
                for j in range(self.grid_size):
                    if self.rewards[i, j] != -5:  # Don't show for obstacles
                        best_action = np.argmax(self.q_table[i, j])
                        ax2.text(j, i, f'{self.action_names[best_action]}\n{q_max[i, j]:.2f}',
                               ha='center', va='center', fontweight='bold', color='white', fontsize=9)
                    else:
                        ax2.text(j, i, 'X', ha='center', va='center', fontweight='bold', color='white', fontsize=12)

            # Plot 3: Learning Curve
            ax3.clear()
            if len(self.rewards_history) > 0:
                ax3.plot(self.rewards_history, color='blue', alpha=0.7, linewidth=1.5)
                if len(self.rewards_history) > 20:
                    window = min(20, len(self.rewards_history) // 4)
                    if window > 1:
                        moving_avg = np.convolve(self.rewards_history, np.ones(window)/window, mode='valid')
                        ax3.plot(range(window-1, len(self.rewards_history)), moving_avg,
                               color='red', linewidth=2, label='Moving Average')
                        ax3.legend()

            ax3.set_title('Learning Progress', fontweight='bold')
            ax3.set_xlabel('Episode')
            ax3.set_ylabel('Total Reward')
            ax3.grid(True, alpha=0.3)

            # Plot 4: Epsilon Decay
            ax4.clear()
            if episode > 0:
                episodes = list(range(episode + 1))
                epsilons = [max(self.epsilon_min, 0.3 * (self.epsilon_decay ** ep)) for ep in episodes]
                ax4.plot(episodes, epsilons, color='green', linewidth=2)
                ax4.axhline(y=self.epsilon_min, color='red', linestyle='--', alpha=0.7, label='Min Epsilon')
                ax4.legend()

            ax4.set_title('Exploration Rate', fontweight='bold')
            ax4.set_xlabel('Episode')
            ax4.set_ylabel('Epsilon')
            ax4.grid(True, alpha=0.3)

            plt.tight_layout()

            if show_plot:
                plt.show()

            return fig

        except Exception as e:
            print(f"Matplotlib error: {e}")
            print("Falling back to text-based visualization...")
            self.print_grid(current_state, path, f"Episode {episode}")
            return None

    def train_episode(self, episode):
        """Train for one episode"""
        state = (0, 0)  # Start position
        total_reward = 0
        steps = 0
        max_steps = 50
        episode_path = [state]
        visited_states = set()  # Track visited states to avoid loops

        while state != (2, 2) and steps < max_steps:
            # Check if we're stuck in a loop
            if state in visited_states and steps > 10:
                # Add small penalty for revisiting states
                total_reward -= 0.05
            visited_states.add(state)

            action = self.get_next_action(state)
            next_state = self.get_next_state(state, action)

            # If action doesn't change state (hitting wall), give small penalty
            if next_state == state:
                reward = -0.1
            else:
                reward = self.rewards[next_state[0], next_state[1]]

            # Don't move into obstacles
            if not self.is_valid_state(next_state):
                reward = -1.0  # Penalty for trying to move into obstacle
                next_state = state  # Stay in current state

            # Q-learning update
            current_q = self.q_table[state[0], state[1], action]
            if next_state != state:  # Only update if we actually moved
                next_max_q = np.max(self.q_table[next_state[0], next_state[1]])
            else:
                next_max_q = 0  # No future reward if we didn't move

            new_q = current_q + self.learning_rate * (reward + self.discount_factor * next_max_q - current_q)
            self.q_table[state[0], state[1], action] = new_q

            if next_state != state:  # Only update state if we actually moved
                state = next_state
                episode_path.append(state)

            total_reward += reward
            steps += 1

            # Early termination if we hit an obstacle
            if not self.is_valid_state(state):
                break

        # Update epsilon
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)
        self.rewards_history.append(total_reward)

        return episode_path, total_reward

    def train(self, episodes=200, visualize_interval=50, use_matplotlib=True):
        """Train the Q-learning agent"""
        print("ü§ñ Starting Q-Learning Training")
        print("=" * 50)

        for episode in range(episodes):
            episode_path, total_reward = self.train_episode(episode)
            self.path_history.append(episode_path)

            # Show progress
            if episode % 25 == 0:
                avg_reward = np.mean(self.rewards_history[-25:]) if len(self.rewards_history) >= 25 else np.mean(self.rewards_history)
                print(f"Episode {episode:3d} | Avg Reward: {avg_reward:6.3f} | Epsilon: {self.epsilon:.3f} | Path Length: {len(episode_path)}")

            # Visualize selected episodes
            if episode % visualize_interval == 0 or episode == episodes - 1:
                if use_matplotlib:
                    try:
                        fig = self.create_visual_plot(episode_path[-1], episode_path, episode, show_plot=False)
                        if fig:
                            plt.pause(0.1)
                            plt.close(fig)
                    except Exception as e:
                        print(f"Visualization error: {e}")

                # Always show text version for key episodes
                if episode % visualize_interval == 0:
                    self.print_grid(episode_path[-1], episode_path, f"Episode {episode} - Robot Path")
                    if episode % (visualize_interval * 2) == 0:
                        self.visualize_q_table_text()

        final_avg = np.mean(self.rewards_history[-20:]) if len(self.rewards_history) >= 20 else np.mean(self.rewards_history)
        print(f"\n‚úÖ Training completed! Final average reward: {final_avg:.3f}")

    def demonstrate_optimal_policy(self, use_matplotlib=True):
        """Demonstrate the learned optimal policy"""
        print("\nüéØ Demonstrating Optimal Policy:")
        print("=" * 40)

        state = (0, 0)
        optimal_path = [state]
        steps = 0
        max_steps = 20
        visited_states = set()

        print(f"Starting at: {state}")

        while state != (2, 2) and steps < max_steps:
            # Avoid infinite loops
            if state in visited_states and steps > 5:
                print("‚ö†Ô∏è  Detected potential loop, breaking...")
                break
            visited_states.add(state)

            # Get best action (no exploration)
            q_values = self.q_table[state[0], state[1]]
            action = np.argmax(q_values)

            # If all Q-values are the same, try a different approach
            if np.all(q_values == q_values[0]):
                print(f"‚ö†Ô∏è  All actions have same Q-value at {state}, using random action")
                action = np.random.randint(4)

            next_state = self.get_next_state(state, action)

            # Don't move into obstacles or invalid states
            if not self.is_valid_state(next_state) or next_state == state:
                # Try other actions
                for alt_action in range(4):
                    if alt_action != action:
                        alt_next_state = self.get_next_state(state, alt_action)
                        if self.is_valid_state(alt_next_state) and alt_next_state != state:
                            action = alt_action
                            next_state = alt_next_state
                            break

            print(f"Step {steps + 1}: {state} ‚Üí {self.action_names[action]} ‚Üí {next_state}")

            if next_state == state:
                print("‚ö†Ô∏è  Robot cannot move, policy may not be optimal")
                break

            state = next_state
            optimal_path.append(state)
            steps += 1

            # Show step-by-step progress
            self.print_grid(state, optimal_path, f"Step {steps}")
            time.sleep(0.5)  # Pause to show movement

        if state == (2, 2):
            print(f"üéâ Goal reached in {steps} steps!")
        else:
            print(f"‚ùå Failed to reach goal within step limit. Final position: {state}")

        print(f"Optimal path: {optimal_path}")

        # Final visualization
        if use_matplotlib:
            try:
                # Pass the episode number as integer
                final_episode = len(self.rewards_history) if self.rewards_history else 0
                self.create_visual_plot(state, optimal_path, episode=final_episode, show_plot=True)
            except Exception as e:
                print(f"Final visualization error: {e}")
                print("Final text visualization:")
                self.print_grid(state, optimal_path, "Final Optimal Policy")
                self.visualize_q_table_text()
        else:
            self.print_grid(state, optimal_path, "Final Optimal Policy")
            self.visualize_q_table_text()

        return optimal_path

    def add_obstacle(self, row, col):
        """Add an obstacle to the environment"""
        if 0 <= row < self.grid_size and 0 <= col < self.grid_size:
            if (row, col) != (0, 0) and (row, col) != (2, 2):
                self.rewards[row, col] = -5
                # Reset Q-values for the obstacle state
                self.q_table[row, col, :] = 0
                print(f"üöß Added obstacle at ({row}, {col})")
                return True
        return False

def main():
    print("üéÆ Q-Learning Robot Navigation Demo")
    print("=" * 40)
    print("Legend:")
    print("üü¶ S = Start position")
    print("üü© G = Goal position")
    print("üü• X = Obstacles")
    print("ü§ñ = Robot position")
    print("Numbers = Path steps")

    # Create the Q-learning robot
    robot = QLearningRobot(grid_size=3)

    print("\n‚ö° Starting training...")

    # Train the robot with more episodes
    robot.train(episodes=300, visualize_interval=50, use_matplotlib=True)

    # Add an obstacle and continue training
    print("\nüöß Adding new obstacle and continuing training...")
    robot.add_obstacle(1, 0)
    robot.train(episodes=150, visualize_interval=25, use_matplotlib=True)

    # Demonstrate optimal policy
    print("\n" + "="*50)
    optimal_path = robot.demonstrate_optimal_policy(use_matplotlib=True)

    return robot

# Run the demo
if __name__ == "__main__":
    robot = main()
```
